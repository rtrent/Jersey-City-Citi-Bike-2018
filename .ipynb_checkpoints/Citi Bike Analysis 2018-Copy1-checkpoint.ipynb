## A) Read files
Load data into single dataframe - we will use this to reformat and clean the data all together before splitting it into two separate dataframes

```python
#create a list of all files for JC in 2017 using file name
files = glob.glob('data/JC-201*.csv')
files
```

```python
df = pd.DataFrame()

for file in files:
    temp = pd.read_csv(file) #load each montly data csv
    temp.rename(str.lower, axis='columns', inplace = True) #standardize column names
    temp.columns = temp.columns.str.replace(" ","")
    df = df.append(temp) #combine into sinlge dataframe

df.reset_index(inplace = True, drop = True) #create a new overall index
```

```python
#confirm that no columns are duplicated based on variations in naming conventions
#for 2017 data this was a problem, but all column names on 2018 appear to be consistent
df.columns
```

## B) Clean data
Clean data and create new variables for later analyses

```python
#we don't need all the extra variables for the later analyses - so removing them to make work a little faster
df.drop(['stoptime', 'startstationid', 'startstationlatitude', 
         'startstationlongitude', 'endstationid', 'endstationlatitude',
        'endstationlongitude', 'usertype', 'bikeid'], axis=1, inplace=True)
df.head()
```

```python
#set data type as datetime to allow for timeseries analyses
df.starttime = pd.to_datetime(df.starttime)
```

```python
#create additional attributes to desribe the time of the Citi Bike trip
df['year'] = df['starttime'].dt.year
df['month'] = df['starttime'].dt.month
df['dayyear'] = df['starttime'].dt.dayofyear
df['dayweek'] = df['starttime'].dt.dayofweek
df['daymonth'] = df['starttime'].dt.day
df['hour'] = df['starttime'].dt.hour
df['date'] = df['starttime'].dt.date
```

```python
#several records have a missing birth year automatically set to 1887, replacing with NAN
df.loc[df['birthyear'] == 1887.0, 'birthyear'] = 0
```

```python
#some birth years are NA, replacing with NAN
df['birthyear'] = df['birthyear'].fillna(0)
```

```python
df['birthyear'] = df['birthyear'].astype(int)
```

```python
df.head()
```

```python
#calculate rough user age
df['age'] = np.where(df['year']==2018, 2018 - df['birthyear'], 2017 - df['birthyear'])
df.head()
```

```python
#helper function to create age category
def age_category(age):
    if (age < 25):
        return '17 - 24'
    elif (age >=25 and age < 35):
        return '25 - 34'
    elif (age >=35 and age < 45):
        return '35 - 44'
    elif (age >=45 and age < 55):
        return '45 - 54'
    elif (age >= 55 and age < 100):
        return '55 and up'
    else:
        return 'other'
```

```python
#apply helper age function and set as type category
df['age_cat'] = df.age.apply(age_category)
df['age_cat'] = df['age_cat'].astype("category")
```

```python
#apply type of category to gender
df['gender'] = df['gender'].astype("category")
```

```python
#split data sets into years
df_18 = df[(df.year == 2018)]
df_17 = df[(df.year == 2017)]
```

```python
#creating a new dataframe with a DateTime index that will allow for plotting all trips over time more easily
bf_18 = pd.DataFrame(df_18.starttime.value_counts())
bf_18.index = pd.to_datetime(bf_18.index)
bf_18.sort_index(inplace = True)
bf_18.rename(columns={'starttime':'count'}, inplace = True)

bf_18.head()
#this shows number of trips that occured by the second
```

```python
#creating a new dataframe with a DateTime index that will allow for plotting all trips over time more easily
bf_17 = pd.DataFrame(df_17.starttime.value_counts())
bf_17.index = pd.to_datetime(bf_17.index)
bf_17.sort_index(inplace = True)
bf_17.rename(columns={'starttime':'count'}, inplace = True)

bf_17.head()
#this shows number of trips that occured by the second
```

```python
#creating crosstab of age and gender intersection
gender_age_17 = pd.crosstab(index=demographics[demographics.year == 2017]['gender'], columns=demographics[demographics.year == 2017]['age_cat'], rownames=['Gender'], colnames=['Age Bucket'])

gender_age_17 = pd.DataFrame(gender_age_17) #saving it as a separate dataframe
gender_age_17 = gender_age_17.rename(index={1: 'Male', 2: 'Female'}) #adding labels for gender rows
gender_age_17 = (gender_age_17 / len(demographics[demographics.year == 2017])).round(2) #rounding data to two decimal places

gender_age_17.to_csv('tables/gender+age_cross17.csv')
#each cell shows what percent of the trips were taken by that age + gender intersection
#numbers are decimals, all cells together sum to 1.00
```

```python
#counting the number of times each station served as the origin of a trip for each month
start_point = df_18.groupby(['startstationname', 'month']).size().reset_index().rename(columns={'startstationname': 'station', 0:'startcount'})
start_point.head()
```

```python
#all start station names are only for stations that are within Jersey City
#this can be used as a list of stations for our final counts
df.startstationname.unique()
```

```python
#counting the number of times each station served as the destination of a trip for each month
end_point = df_18.groupby(['endstationname', 'month']).size().reset_index().rename(columns={'endstationname': 'station', 0:'endcount'})
end_point = end_point[end_point['station'].isin(df.startstationname.unique())].reset_index(drop = True)
end_point.head()
```

```python
#combining the number of times each station was used as an origin or destination for a Citi Bike trip
monthly_station_use = pd.merge(start_point, end_point, on=['station', 'month'])

#calculating a total use metric
monthly_station_use['total'] = monthly_station_use['endcount'] + monthly_station_use['startcount']

monthly_station_use
```

```python
#crosstab of the total number of times each station was used by month
montly_cross = pd.crosstab(index = monthly_station_use.station, columns = monthly_station_use.month, values = monthly_station_use.total, aggfunc = 'sum')
montly_cross.to_csv('tables/monthly_cross.csv')
montly_cross
#JCBS Depot should be ignored - seems like some kind of management location
```

```python
#NaNs represent months where a station was not active
#Leaving these as NaNs will make sure that they are not part of the calcualtion when taking the mean#
#Several stations were closed / opened during the course of the year.
#Seems like those that were closed were less used and the newer opened ones became popular very quickly
```

```python
#average for each citi bike station by number of months
montly_means = montly_cross.mean(axis = 1).reset_index().rename(columns = {0:'avg trips'}).sort_values('avg trips', ascending = False).reset_index(drop = True)
montly_means.to_csv('tables/montlymean.csv')
```
