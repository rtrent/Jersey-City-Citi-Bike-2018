```python
#create a list of all files for JC in 2017 using file name
files = glob.glob('data/JC-2017*.csv')
files
```

```python
df = pd.DataFrame()

for file in files:
    temp = pd.read_csv(file) #load each montly data csv
    temp.rename(str.lower, axis='columns', inplace = True) #standardize column names
    temp.columns = temp.columns.str.replace(" ","")
    df = df.append(temp) #combine into sinlge dataframe

df.reset_index(inplace = True) #create a new overall index
```

```python
#set data type as datetime to allow for timeseries analyses
df.starttime = pd.to_datetime(df.starttime)
df.stoptime = pd.to_datetime(df.stoptime)
```

```python
#several records have a missing birth year automatically set to 1887, replacing with 0
df.loc[df['birthyear'] == 1887.0, 'birthyear'] = 0.0
```

```python
#some birth years are NA, replacing with 0
df['birthyear'] = df['birthyear'].fillna(0.0).astype(int)
```

```python
#calculate rough user age
df['age'] = 2017 - df['birthyear'] 
```

```python
#helper function to create age category
def age_category(age):
    if (age < 25):
        return '17 - 24'
    elif (age >=25 and age < 35):
        return '25 - 34'
    elif (age >=35 and age < 45):
        return '35 - 44'
    elif (age >=45 and age < 55):
        return '45 - 54'
    elif (age >= 55 and age < 120):
        return '55 and up'
    else:
        return 'other'
```

```python
#apply helper age function and set as type category
df['age_cat'] = df.age.apply(age_category)
df['age_cat'] = df['age_cat'].astype("category")
```

```python
#apply type of category to gender
df['gender'] = df['gender'].astype("category")
```

```python
#create additional attributes to desribe the time of the Citi Bike trip
df['month'] = df['starttime'].dt.month
df['dayyear'] = df['starttime'].dt.dayofyear
df['dayweek'] = df['starttime'].dt.dayofweek
df['daymonth'] = df['starttime'].dt.day
df['hour'] = df['starttime'].dt.hour
df['date'] = df['starttime'].dt.date
```

```python
#list of federal holidays to remove from weekday data (federal holidays are all on weekdays)
holiday_times = []

holidays = ['01-02', #new years day
           '01-16', #mlk
           '02-20', #pres day
           '05-29', #memorial day
           '07-04', #july 4th
           '09-04', #labor day
           '11-10', #veterans day
           '11-23', #thanksgiving
           '12-25'] #xmas
#did not do Columbus day as many companies no longer give off for that

#loop to create an array of DateTime indexes for all hours in these holidays
for holiday in holidays:
    for i in range(0,24):
        holiday_times.append(pd.Timestamp('2017-' + holiday + ' %d:00:00' %i))

#first 5 elements of holiday_times
holiday_times[:5]
```

```python
#counting the number of times each station served as the origin of a trip for each month
start_point = df.groupby(['startstationname', 'month']).size().reset_index().rename(columns={'startstationname': 'station', 0:'startcount'})
start_point.head()
```

```python
#all start station names are only for stations that are within Jersey City
#this can be used as a list of stations for our final counts
df.startstationname.unique()
```

```python
#counting the number of times each station served as the destination of a trip for each month
end_point = df.groupby(['endstationname', 'month']).size().reset_index().rename(columns={'endstationname': 'station', 0:'endcount'})
end_point = end_point[end_point['station'].isin(df.startstationname.unique())].reset_index(drop = True)
end_point.head()
```

```python
#combining the number of times each station was used as an origin or destination for a Citi Bike trip
monthly_station_use = pd.merge(start_point, end_point, on=['station', 'month'])

#calculating a total use metric
monthly_station_use['total'] = monthly_station_use['endcount'] + monthly_station_use['startcount']

monthly_station_use
```

```python
#crosstab of the total number of times each station was used by month
montly_cross = pd.crosstab(index = monthly_station_use.station, columns = monthly_station_use.month, values = monthly_station_use.total, aggfunc = 'sum')
montly_cross.to_csv('tables/monthly_cross.csv')
montly_cross
#JCBS Depot should be ignored - seems like some kind of management location
```

```python
#NaNs represent months where a station was not active
#leaving these as NaNs will make sure that they are not part of the calcualtion when taking the mean
#some stations legitimately had 0 uses in a single month
#replacing values to account for this
montly_cross.loc['Bethune Center'][11] = 0
montly_cross.loc['Bayside Park'][1] = 0
```

```python
#average for each citi bike station by number of months
montly_means = montly_cross.mean(axis = 1).reset_index().rename(columns = {0:'avg trips'}).sort_values('avg trips', ascending = False).reset_index(drop = True)
montly_means.to_csv('tables/montlymean.csv')
```

```python
#rebalancing occurs when Motivate employees take bikes at station A and move them to station B
#this happens in cases where there are not enough bikes to accomadate riders at station A or if there are not enough empty docs at station B
#rebalancing is not explicitly called out in the data but it can be calculated

#if BIKE#1 is ridden from Station A to Station B, then the next time BIKE#1 shows up in the data it should be checked out from Station B
#in some cases we see bikes ridden from Station A to Station B, and then next time they show up in data, they are checked out of Station C
#in order to get to station C without a new trip record being created, they would have to be moved by Motivate

#note - this is slow an inefficient but it works

rebalance = pd.DataFrame()

for bike in bikes:
    #get the indexes of all the rides each bike took
    events = df.index[df['bikeid'] == bike]

    #repeat for each ride the bike took
    for x in range(0,len(events)):
        #find the end station to the first ride
        end = df['endstationname'][df['bikeid'] == bike][events[x]]
        if x+1 < len(events):
            #find the start station to the ride right after
            start = df['startstationname'][df['bikeid'] == bike][events[x+1]]
            
            if end != start:
                #print(False)
                d = {'bikeid': [bike], 'rebal from': [end], 'rebal to': [start]}
                rebalevent = pd.DataFrame(data = d, dtype='object')
                rebalance = rebalance.append(rebalevent)
```

```python
#crosstab of all instances of bikes moving between stations
pd.crosstab(rebalance['rebal from'], rebalance['rebal to'])
```

```python
#count of times bikes were rebalanced between combinations of stations
rebalance_list = rebalance.groupby(['rebal from','rebal to']).size().reset_index().rename(columns={0:'count'}).sort_values('count', ascending = False).reset_index(drop = True)
rebalance_list.head(25).to_csv('tables/rebalance_list.csv')
rebalance_list.head(25) #show the top 25 rebalances
```
