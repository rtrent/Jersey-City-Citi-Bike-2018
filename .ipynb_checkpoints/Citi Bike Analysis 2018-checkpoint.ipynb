## A) Read files
Load data into single dataframe - we will use this to reformat and clean the data all together before splitting it into two separate dataframes

```python
#create a list of all files for JC in 2017 using file name
files = glob.glob('data/JC-201*.csv')
files
```

```python
df = pd.DataFrame()

for file in files:
    temp = pd.read_csv(file) #load each montly data csv
    temp.rename(str.lower, axis='columns', inplace = True) #standardize column names
    temp.columns = temp.columns.str.replace(" ","")
    df = df.append(temp) #combine into sinlge dataframe

df.reset_index(inplace = True, drop = True) #create a new overall index
```

```python
#confirm that no columns are duplicated based on variations in naming conventions
#for 2017 data this was a problem, but all column names on 2018 appear to be consistent
df.columns
```

```python
#initial look at the data
df.tail()
```

## B) Clean data
Clean data and create new variables for later analyses

```python
#we don't need all the extra variables for the later analyses - so removing them to make work a little faster
df.drop(['stoptime', 'startstationid', 'startstationlatitude', 
         'startstationlongitude', 'endstationid', 'endstationlatitude',
        'endstationlongitude', 'usertype', 'bikeid'], axis=1, inplace=True)
df.head()
```

```python
#set data type as datetime to allow for timeseries analyses
df.starttime = pd.to_datetime(df.starttime)
```

```python
#create additional attributes to desribe the time of the Citi Bike trip
df['year'] = df['starttime'].dt.year
df['month'] = df['starttime'].dt.month
df['dayyear'] = df['starttime'].dt.dayofyear
df['dayweek'] = df['starttime'].dt.dayofweek
df['daymonth'] = df['starttime'].dt.day
df['hour'] = df['starttime'].dt.hour
df['date'] = df['starttime'].dt.date
```

```python
#several records have a missing birth year automatically set to 1887, replacing with NAN
df.loc[df['birthyear'] == 1887.0, 'birthyear'] = 0
```

```python
#some birth years are NA, replacing with NAN
df['birthyear'] = df['birthyear'].fillna(0)
```

```python
df['birthyear'] = df['birthyear'].astype(int)
```

```python
df.head()
```

```python
#calculate rough user age
df['age'] = np.where(df['year']==2018, 2018 - df['birthyear'], 2017 - df['birthyear'])
df.head()
```

```python
#helper function to create age category
def age_category(age):
    if (age < 25):
        return '17 - 24'
    elif (age >=25 and age < 35):
        return '25 - 34'
    elif (age >=35 and age < 45):
        return '35 - 44'
    elif (age >=45 and age < 55):
        return '45 - 54'
    elif (age >= 55 and age < 100):
        return '55 and up'
    else:
        return 'other'
```

```python
#apply helper age function and set as type category
df['age_cat'] = df.age.apply(age_category)
df['age_cat'] = df['age_cat'].astype("category")
```

```python
#apply type of category to gender
df['gender'] = df['gender'].astype("category")
```

```python
#split data sets into years
df_18 = df[(df.year == 2018)]
df_17 = df[(df.year == 2017)]
```

```python
#creating a new dataframe with a DateTime index that will allow for plotting all trips over time more easily
bf_18 = pd.DataFrame(df_18.starttime.value_counts())
bf_18.index = pd.to_datetime(bf_18.index)
bf_18.sort_index(inplace = True)
bf_18.rename(columns={'starttime':'count'}, inplace = True)

bf_18.head()
#this shows number of trips that occured by the second
```

```python
#creating a new dataframe with a DateTime index that will allow for plotting all trips over time more easily
bf_17 = pd.DataFrame(df_17.starttime.value_counts())
bf_17.index = pd.to_datetime(bf_17.index)
bf_17.sort_index(inplace = True)
bf_17.rename(columns={'starttime':'count'}, inplace = True)

bf_17.head()
#this shows number of trips that occured by the second
```

```python
#designating days of the week as weekends or weekdays to enable analyses that compare the two
weekday_type = np.where(bf_18.index.weekday < 5, 'Weekday', 'Weekend')
weekday_type
```

```python
#data for just weekdays (M-F)
hourly_week = bf_18[weekday_type == 'Weekday'].resample('H').sum()

#averages for weekday times (including holidays)
hourly_week.groupby(hourly_week.index.hour).sum() / weekday_18['count'][:5].sum()
#total sum of trips at each our for weekdays divided by the number of weekdays (M-F) in 2017, giving the average 
```

```python
#data for just weekends (Sa+Su)
hourly_weekend = bf_18[weekday_type == 'Weekend'].resample('H').sum()

hourly_weekend.groupby(hourly_weekend.index.hour).sum() / weekday_18['count'][5:].sum()
#total sum of trips at each our for weekends divided by the number of weekends (Sa+Su) in 2017, giving the average
```

```python
#list of federal holidays to remove from weekday data (federal holidays are all on weekdays)
holiday_times = []
            
holidays = [#'2017-01-02', #new years day
            #'2017-01-16', #mlk
            #'2017-02-20', #pres day
            #'2017-05-29', #memorial day
            #'2017-07-04', #july 4th
            #'2017-09-04', #labor day
            #'2017-11-10', #veterans day
            #'2017-11-23', #thanksgiving
            #'2017-12-25'] #xmas
            '2018-01-01', #new years day
            '2018-01-15', #mlk
            '2018-02-19', #pres day
            '2018-05-28', #memorial day
            '2018-07-04', #july 4th
            '2018-09-03', #labor day
            '2018-11-12', #veterans day
            '2018-11-22', #thanksgiving
            '2018-12-25'] #xmas
#did not do Columbus day as many companies no longer give off for that

#loop to create an array of DateTime indexes for all hours in these holidays
for holiday in holidays:
    for i in range(0,24):
        holiday_times.append(pd.Timestamp(holiday + ' %d:00:00' %i))

#first 5 elements of holiday_times
holiday_times[:5]
```

```python
#removing holidays from weekday
hourly_week.drop(holiday_times, errors = 'ignore', inplace = True)
```

```python
#averages by time of day for weekdays (excludes holidays)
hourly_week.groupby(hourly_week.index.hour).sum() / (weekday_18['count'][:5].sum() - len(holidays))
#total of trips by hours of day for non-holiday/working weekdays (M-F) divided by the total number of working weekdays
```

```python
#plot of average trips by hour of day for non-holiday weekdays and weekends together

#setting data
a = hourly_week.groupby(hourly_week.index.hour).sum() / (weekday_18['count'][:5].sum() - len(holidays))
b = hourly_weekend.groupby(hourly_weekend.index.hour).sum() / weekday_18['count'][5:].sum()


#plotting lines
plt.plot(a, 'r')
plt.plot(b, 'purple')

plt.legend(['work week', 'weekend'], fontsize=12)

#axes
ax = plt.gca()
plt.xticks(range(0,23,2), ['00:00', '02:00', '04:00', '06:00', '08:00', '10:00', '12:00', '14:00', '16:00',
                   '18:00', '20:00', '22:00'], fontsize=10)
plt.xticks(rotation=45)
ax.set_xlim(0, 23)
ax.set_ylim(0, 180)

plt.tick_params(axis='both', which='major', labelsize=12)

#labels
plt.title('Average Trips by Time of Day 2018', fontsize=18)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Average Trips', fontsize=14)

plt.tight_layout()

plt.savefig('charts/trips_timeofday2018.png')
```
